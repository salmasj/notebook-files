```python
# Import data.
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt


# Membaca data dari setiap sheet.
data_frame_1 = pd.read_excel("data pengkom.xlsx", sheet_name="Sheet1")
data_frame_2 = pd.read_excel("data pengkom.xlsx", sheet_name="Sheet2")
data_frame_3 = pd.read_excel("data pengkom.xlsx", sheet_name="Sheet3")
```

2. Dari data pada Sheet 1, buat horizontal bar chart tiap harga makanan.


```python
# Horizontal bar chart untuk menampilkan harga dari setiap makanan
data_frame_1.plot(kind="barh",x="nama",y="harga",title="Harga Setiap Makanan", color="turquoise")
plt.show()
```

3. Dari data pada Sheet 2, buat histogram tinggi badan dengan kelompok tinggi 100-110,
110-120, 120-130, 130-140, 140-150, 150-160.

```python
# Histogram tinggi badan berdasarkan kelompok tinggi: 100-110; 110-120; 120-130; 130-140; 140-150; 150-160. 
data_frame_2[["tinggi_anak"]].plot(kind="hist", bins=[100,110,120,130,140,150,160], rwidth=0.8, color="lightcoral", title="Frekuensi Tinggi Badan")
plt.show()
```

4. Dari data pada Sheet 1, buat pie chart komposisi banyaknya makanan pada tiap jenis.

```python
# Komposisi banyaknya makanan pada tiap jenis
mycolors = ["lightskyblue", "greenyellow", "lightsalmon"]
data_frame_1["jenis"].value_counts().plot(kind = "pie", colors=mycolors, autopct='%1.1f%%', title="Persentase Jenis Makanan")
plt.show()
```

5. Dari data pada Sheet 3, buat line chart pertumbuhan harga.


```python
# Pertumbuhan harga emas dari tahun ke tahun dalam line chart
data_frame_3.plot(kind="line",x="date",y="gold_price", color="lightgreen", title="Pertumbuhan Harga Emas")
plt.show()
```
