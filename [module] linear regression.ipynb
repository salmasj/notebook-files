```python
# import data dan library yang akan digunakan
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df = pd.read_csv('data.csv')
```

## Pengujian Data dan Analisis Visual

### Scatter Plot

```python
# Scatter plot price terhadap horsepower
plt.scatter(df['horsepower'], df['price'])

# Legenda dan judul dari plot
plt.xlabel('Horsepower') # Legenda untuk sumbu-x sebagai ‘Horsepower’
plt.ylabel('Price')      # Legenda untuk sumbu-y sebagai ‘Price’
plt.title('Scatter Plot Horsepower vs Price')  # Judul dari plot
plt.show()   # Tampilkan plot
```

```python
# Box plot horsepower
plt.boxplot(df['horsepower'])
plt.title('Box Plot Horsepower')  # Judul dari plot
plt.show()   # Tampilkan plot
```

```python
# Box plot price
plt.boxplot(df['price'])
plt.title('Box Plot Horsepower')  # Judul dari plot
plt.show()   # Tampilkan plot
```

## Membangun Model Regresi Linear

```python
# Mendefinisikan variabel x yaitu horsepower dan variabel y yaitu price
x = df['horsepower'].values.reshape(-1,1)
y = df['price'].values.reshape(-1,1)

# Membangun model regresi linier
model = LinearRegression().fit(x,y)

# Mendapatkan nilai a dan b pada model y = a + bx
print("a:", model.intercept_) # nilai a atau intercept
print("b: ", model.coef_)     # nilai b atau koefisien slope/gradien
```

```python
type(df['horsepower'])
```

## Memvisualisasikan Regresi Linear Sederhana dengan Grafik

```python
# Plot model regresi linier dengan data
regresi = model.predict(x)

plt.scatter(x, y) # scatter plot dari data asli
plt.plot(x, regresi, color='red') # garis regresi
plt.xlabel('Horsepower') # Legenda untuk sumbu-x sebagai ‘Horsepower’
plt.ylabel('Price')      # Legenda untuk sumbu-y sebagai ‘Price’
plt.title('Plot Horsepower vs Price')  # Judul dari plot
```

## Diagnostik Regresi Linear Sederhana

```python
# Mendapatkan nilai R-squared
print("R-squared: ", model.score(x, y))
```

## Memprediksi Nilai Respons Menggunakan Model Regresi

```python
# Menggunakan model untuk melakukan prediksi price 
# untuk horsepower 220 dan 230
y_prediksi = model.predict([[220], [230]])
print(y_prediksi)
```
