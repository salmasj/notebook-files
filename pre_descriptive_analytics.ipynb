```python
import pandas as pd
data_frame = pd.read_csv('pokemon.csv')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows' , None)
data_frame.head()
```

```python
# 1. Banyaknya (panjangnya) data
print(len(data_frame))
```

```python
# 2. Data ke-100 hingga ke-120 (indeks 99 hingga indeks 119)
data_frame[99:120]
```

```python
# 3. Data pada kolom total_points sampai dengan base_experience
data_frame.loc[ : , 'total_points':'base_experience']
```

```python
# 4. Data pokemon dengan name Pikachu
data_frame.loc[data_frame['name'] == 'Pikachu']
```

```python
# 5.a. Data dengan height maksimum 
height_maksimum = data_frame['height'].idxmax()
data_frame[height_maksimum:height_maksimum+1]
```

```python
# 5.b. Data dengan height minimum 
height_minimum = data_frame['height'].idxmin()
data_frame[height_minimum:height_minimum+1]
```

```python
# 6. Jangkauan dari kolom weight
maksimum = data_frame.max()
minimum = data_frame.min()
range = maksimum['weight'] - minimum['weight']
range
```

```python
# 7. Data yang diurutkan berdasarkan kolom total_points secara menaik
data_frame.sort_values(['total_points'], ascending=[1])
```

```python
# 8. Tabel frekuensi munculnya nilai kolom status
data_frame['status'].value_counts()
```

```python
# 9. Kelompokkan data berdasarkan kolom status dan type, kemudian tampilkan rata-rata kolom attack-nya
data_frame.groupby(['status', 'type'])['attack'].mean()
```

```python
# 10. Rata-rata dari kolom hp
data_frame['hp'].mean()
```

```python
# 11. Pandang kolom catch_rate, base_friendship, dan base_experience. Dengan kolom manakah catch_rate berkorelasi? Berapa koefisien korelasinya?
data_frame['catch_rate'].corr(data_frame['base_experience'])
```
